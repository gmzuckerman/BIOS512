````r
options(repr.plot.res = 300, repr.plot.width = 8, repr.plot.height = 3)

p = ggplot(penguins_long, aes(x = species, y = value, color = sex)) 

p = p + geom_point(position = position_jitter(height = 0.5, width = 0.3), size = 1, alpha = 0.8)

p = p + facet_wrap(~name, nrow = 1, scales="free_y")

p = p + theme(axis.text.x = element_text(hjust = 1, angle = 45))

p = p + stat_summary(fun = median, 
                     fun.min = median, 
                     fun.max = median, 
                     geom = "crossbar", 
                     width = 0.75, 
                     size = 0.2,
                     mapping = aes(group = species))

p
````
